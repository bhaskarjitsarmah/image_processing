```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = plt.imread('lenna.png', 1)
gamma = 1.5
fullRange = np.arange(0, 256)
lut = np.uint8( 255 * np.power( (fullRange / 255.0), gamma) )
output = cv2.LUT(img, lut)
combined = np.hstack([img, output])
```

```python
plt.imshow(combined)
plt.show()
```
