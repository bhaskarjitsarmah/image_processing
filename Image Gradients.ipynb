```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = plt.imread('lenna.png')

sobelx = cv2.Sobel(img, cv2.CV_32F, 1, 0)
sobely = cv2.Sobel(img, cv2.CV_32F, 0, 1)

cv2.normalize(sobelx, dst = sobelx, alpha = 0, beta = 1, norm_type = cv2.NORM_MINMAX, dtype = cv2.CV_32F)
cv2.normalize(sobely, dst = sobely, alpha = 0, beta = 1, norm_type = cv2.NORM_MINMAX, dtype = cv2.CV_32F)
```

```python
combined = np.hstack([img, sobelx, sobely])
plt.imshow(combined)
```
