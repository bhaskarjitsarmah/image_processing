```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = plt.imread('lenna.png', 1)
beta = 100
ycbImg = cv2.cvtColor(img, cv2.COLOR_BGR2YCrCb)
ycbImg = np.float32(ycbImg)
Ychannel, Cr, Cb = cv2.split(ycbImg)

Ychannel = np.clip(Ychannel + beta, 0, 255)
ycbImg = np.uint8(cv2.merge([Ychannel, Cr, Cb]))
imBright = cv2.cvtColor(ycbImg, cv2.COLOR_YCrCb2BGR)

combined = np.hstack([img, imBright])
```

```python
plt.imshow(combined)
plt.show()
```
