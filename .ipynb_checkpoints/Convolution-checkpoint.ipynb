```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Image Smoothing

```python
img = plt.imread('lenna.png', 1)
kernel_size = 5
kernel = np.ones((kernel_size, kernel_size), dtype=np.float32) / kernel_size**2
result = cv2.filter2D(img, -1, kernel, (-1, -1), delta=0, borderType=cv2.BORDER_DEFAULT)
combined = np.hstack([img, result])
```

```python
plt.imshow(combined)
plt.show()
```

### Image Blurring

```python
dst1 = cv2.blur(img, (5,5), (-1, -1))
dst2 = cv2.blur(img, (9,9), (-1, -1))
combined = np.hstack([dst1, dst2])

plt.imshow(combined)
plt.show()
```
