```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Intensity Scaling

```python
img = plt.imread('lenna.png', 1)
scaleFactor = 2
ycbImg = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
ycbImg = np.float32(ycbImg)
Ychannel, Cr, Cb = cv2.split(ycbImg)
Ychannel = np.clip(Ychannel * scaleFactor , 0, 255)
ycbImage = np.uint8( cv2.merge([Ychannel, Cr, Cb]) )
imcontrast = cv2.cvtColor(ycbImage, cv2.COLOR_YCrCb2BGR)
combined = np.hstack([img,imcontrast])
```

```python
plt.imshow(combined)
plt.show()
```

### Histogram Equalization

```python
img = plt.imread('lenna.png', 1)
ycbImg = cv2.cvtColor(img,cv2.COLOR_BGR2YCrCb)
Ychannel, Cr, Cb = cv2.split(ycbImg)
Ychannel = cv2.equalizeHist(Ychannel)
ycbImage = cv2.merge([Ychannel, Cr, Cb])
imcontrast = cv2.cvtColor(ycbImage, cv2.COLOR_YCrCb2BGR)
combined = np.hstack([img,imcontrast])
```

```python
plt.imshow(combined)
plt.show()
```
