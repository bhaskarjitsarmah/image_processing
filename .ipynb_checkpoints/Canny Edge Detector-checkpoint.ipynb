```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
src = cv2.imread('lenna.png', 0)
```

```python
lowThreshold = 50
highThreshold = 100
maxThreshold = 1000

apertureSizes = [3, 5, 7]
maxapertureIndex = 2
apertureIndex = 0

blurAmount = 0
maxBlurAmount = 20
```

```python
def applyCanny():
    if(blurAmount>0):
        blurredSrc = cv2.GaussianBlur(src, (2*blurAmount + 1, 2*blurAmount + 1), 0)
    else:
        blurredSrc = src.copy()
        
    apertureSize = apertureSizes[apertureIndex]
    edges = cv2.Canny(blurredSrc, lowThreshold, highThreshold, apertureSize=apertureSize)
    plt.imshow(edges)
    plt.show()
```

```python
applyCanny()
```
