```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = plt.imread('lenna.png', 1)

sharpen = np.array(([0, -1, 0], [-1, 5, -1], [0, -1, 0]), dtype='int')
sharpenOutput = cv2.filter2D(img, -1, sharpen)
combined = np.hstack([img, sharpenOutput])

plt.imshow(combined)
plt.show()
```
